## 2.1 Реалізація DFS на основі ітерацій

```python
from collections import deque

class DepthFirstSearch:
    def __init__(self):
        self.stack = deque()

    def dfs(self, vertex_list):
        for vertex in vertex_list:
            if not vertex.visited:
                vertex.visited = True
                self.dfs_helper(vertex)

    def dfs_helper(self, root_vertex):
        self.stack.append(root_vertex)
        root_vertex.visited = True

        while self.stack:
            actual_vertex = self.stack.pop()
            print(actual_vertex)

            for neighbor in actual_vertex.adjacency_list:
                if not neighbor.visited:
                    neighbor.visited = True
                    self.stack.append(neighbor)
```

```python
from lib.helpers import visualize_graph
from lib.classes import Graph

edges = [
    ("A", ["B", "F", "G"]),
    ("B", ["C", "D"]),
    ("G", ["H"]),
    ("D", ["E"])
]

# Create graph with edges
graph = Graph(edges, directed=False)

# Visualize the graph
visualize_graph(graph)

# Perform DFS traversal
dfs = DepthFirstSearch()
vertices = list(graph.vertices.values())
dfs.dfs(vertices)

# Reset visits for potential further use
graph.reset_visits() 
```

## 2.2 Реалізація DFS на базі рекурсії

```python
class DepthFirstSearchRecursion:
    def dfs(self, vertex_list):
        # if we have independent clusters
        for vertex in vertex_list:
            if not vertex.visited:
                vertex.visited = True
                self.dfs_helper(vertex)

    def dfs_helper(self, vertex):
        print(vertex)

        for neighbor in vertex.adjacency_list:
            if not neighbor.visited:
                neighbor.visited = True
                # calling recursively this method
                self.dfs_helper(neighbor)
```

```python
edges = [
    ("A", ["B"]),
    ("B", ["C", "D"]),
    ("D", ["E"])
]

graph = Graph(edges, directed=False)
visualize_graph(graph)

dfs = DepthFirstSearchRecursion()
vertices = list(graph.vertices.values())
dfs.dfs(vertices)

graph.reset_visits()
```
